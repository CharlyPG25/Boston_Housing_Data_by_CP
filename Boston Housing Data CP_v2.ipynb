## Discretize the age variable into three groups of 35 years and younger, between 35 and 70 years and 70 years and older

```python
# Define bins and labels
bins = [0, 35, 70, float('inf')]
labels = ['<=35', '35-70', '70+']

# Discretize AGE variable
boston_df['AGE_GROUP'] = pd.cut(boston_df['AGE'], bins=bins, labels=labels, right=True)

# Show first rows
print(boston_df[['AGE', 'AGE_GROUP']].head(10))

# Provide a boxplot for the MEDV variable vs the AGE variable. 

sns.boxplot(x='MEDV', y='AGE_GROUP', data=boston_df)
pyplot.title('Median Value of Owner-Occupied Homes vs Age Group')
pyplot.xlabel('Median Value of Owner-Occupied Homes')
pyplot.ylabel('Age Group')
pyplot.show()
```

## I there a significant difference in median value of houses bounded by the Charles river or not?

```python
# Split groups based on CHAS
medv_river = boston_df[boston_df['CHAS'] == 1]['MEDV']
medv_no_river = boston_df[boston_df['CHAS'] == 0]['MEDV']

# Perform independent t-test
t_stat, p_val = stats.ttest_ind(medv_river, medv_no_river, equal_var=False)  # Welchs t-test
print("T-statistic:", t_stat)
print("P-value:", p_val)
```

Because p-value = 0.0036 is less than alpha = 0.05, we reject the null hypotesis. So there is a statistically significant difference in the median values of houses by the Charles River or not. Houses located next to the Charles River have significantly higher median values.

## Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

```python
# Perform ANOVA test
# Define the formula for the ANOVA test
formula = 'MEDV ~ C(AGE_GROUP)'

# Create the OLS model
lm = ols(formula, data=boston_df).fit()

# Perform ANOVA and print the table
anova_table = sm.stats.anova_lm(lm)
print(anova_table)
```

Since p-value = 1.71e-15 is less than alpha = 0.05, we reject the null hypotesis. So there is a statistically significant difference in the median values of houses for each proportion of owner occupied built prior to 1940. The median values go down when houses are elders.

## Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town?

```python
# Perform Pearson correlation test
correlation, p_value = stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])

print(f"Pearson Correlation Coefficient: {correlation}")
print(f"P-value: {p_value}")
```

The Pearson Correlation Coefficient is approximately 0.764, which indicates a strong positive linear relationship between Nitric oxide concentrations (NOX) and the proportion of non-retail business acres per town (INDUS).

The p-value is approximately $7.91 \times 10^{-98}$, which is much less than the common significance level of 0.05.

Therefore, we reject the null hypothesis and conclude that there is a statistically significant linear relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town. We cannot conclude that there is no relationship.

## What is the impact of an additional weighted distance to the five Boston employment centres on the median value of owner occupied homes? 

```python
# Perform regression analysis
# Define the independent and dependent variables
X = boston_df['DIS']
y = boston_df['MEDV']

# Add a constant to the independent variable for the intercept
X = sm.add_constant(X)

# Create and fit the OLS model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```

The coefficient for DIS is approximately 1.0916. This means that for every one-unit increase in the weighted distance to the five Boston employment centers, the median value of owner-occupied homes is estimated to increase by approximately 1091.60 (since MEDV is in $1000's).

The p-value for DIS is 0.000, which is much less than 0.05. This indicates that the weighted distance to employment centers is a statistically significant predictor of the median value of owner-occupied homes.

The R-squared value is 0.062, which means that approximately 6.2% of the variation in the median value of owner-occupied homes can be explained by the weighted distance to the five Boston employment centers. This suggests that DIS is a statistically significant predictor, but it only explains a small portion of the variance in MEDV, indicating that other factors also influence housing prices.
